```python
import os
from os import listdir
import dice_score as ds
import cv2
```

```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
```

```python
BBBC_clustered = "../cell_clustering/BBBC_Output/"
cluster_list = load_images_from_folder(BBBC_clustered)
```

```python
BBBC_gt = "../data/human_ht29_colon_cancer_2_foreground/"
gt_list = load_images_from_folder(BBBC_gt)
```

```python
for i in range(len(cluster_list)):
    print(os.listdir(BBBC_clustered)[i])
    print(os.listdir(BBBC_gt)[int(i/(len(cluster_list)/len(gt_list)))])
    ds.dice_foreground(gt_list[int(i/(len(cluster_list)/len(gt_list)))], cluster_list[i])
```
