Intensity clipping: Reducing the intensity spectrum of the images.

```python
T1 = 50
T2 = 100
```

```python
figure(figsize = (15, 8))
imshow(yeast.clip(T1, T2))
colorbar()
```

```python
figure(figsize = (15, 8))
imshow(nuclei.clip(T1, T2))
colorbar()
```
